# Visualization Notebook

This notebook contains all the plots and visualizations extracted from the EDA notebook. The purpose is to organize and execute all plotting-related code in one place for better clarity and analysis.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# Define a function to plot histograms
def plot_histogram(df, column_name):
    plt.figure(figsize=(5, 3))
    sns.histplot(df[column_name], kde=True)
    plt.title(f"Distribution of {column_name}")
    col_mean = df[column_name].mean()
    col_median = df[column_name].median()
    plt.axvline(col_mean, color="red", linestyle="--", label="Mean")
    plt.axvline(col_median, color="green", linestyle="-", label="Median")
    plt.legend()
    plt.show()

# Example usage
plot_histogram(df, "tenure")
plot_histogram(df, "MonthlyCharges")
plot_histogram(df, "TotalCharges")
```

```python
# Define a function to plot scatter plots
def plot_scatterplot(df, x_column, y_column):
    plt.figure(figsize=(7, 4))
    sns.scatterplot(x=df[x_column], y=df[y_column])
    plt.title(f"Scatter Plot of {x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Example usage
plot_scatterplot(df, "tenure", "MonthlyCharges")
plot_scatterplot(df, "tenure", "TotalCharges")
```

```python
# Plot a correlation heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(df[["tenure", "MonthlyCharges", "TotalCharges"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
